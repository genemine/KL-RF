```python
from sklearn.svm import SVC
import numpy as np
import matplotlib.pyplot as plt
from sklearn.ensemble import RandomForestClassifier
from sklearn import metrics
from sklearn.model_selection import train_test_split
from sklearn.model_selection import GridSearchCV
import numpy as np
```

```python
X_ppi = np.load("E:/科研/KL-RF/data/result/X_raw_ppi.npy")
X_go = np.load("E:/科研/KL-RF/data/result/X_raw_go.npy")
X_joint = np.load("E:/科研/KL-RF/data/result/X_raw_joint.npy")
y = np.load("E:/科研/KL-RF/data/result/y_train_ppi.npy")
X = np.load("E:/科研/KL-RF/data/result/X.npy")
print(X_ppi.shape)
print(X_go.shape)
print(X.shape)
```

```python
#ppi
#默认参数
X_train_ppi, X_test_ppi, y_train_ppi, y_test_ppi = train_test_split(X_ppi, y, test_size=0.8)

rf0 = RandomForestClassifier(oob_score=True, random_state=10)
rf0.fit(X_train_ppi, y_train_ppi)
y_pred = rf0.predict(X_test_ppi)

print('Accuracy : %f' % metrics.accuracy_score(y_test_ppi, y_pred))
print('AUC : %f' % metrics.roc_auc_score(y_test_ppi, y_pred))
print('Precision : %f' % metrics.precision_score(y_test_ppi, y_pred))
print('Recall : %f' % metrics.recall_score(y_test_ppi, y_pred))
print('Average_precision : %f' % metrics.average_precision_score(y_test_ppi, y_pred))
```

```python
param_test1 = {'n_estimators': range(150, 501, 50), 'max_depth': range(3, 31, 3), 'min_samples_split': range(1, 10, 2)}
gsearch1 = GridSearchCV(estimator=RandomForestClassifier(min_samples_leaf=20, max_features='sqrt', random_state=10),
                        param_grid=param_test1, scoring='roc_auc', cv=5)
gsearch1.fit(X_train_ppi, y_train_ppi)
print(gsearch1.best_params_)
print(gsearch1.best_score_)
```

```python
param_test2 = {'min_samples_split': range(10, 151, 20),'max_features': range(1, 12, 2)}
gsearch2 = GridSearchCV(estimator=RandomForestClassifier(n_estimators=400, max_depth=6, min_samples_split=3,
                                                         oob_score=True, random_state=10),
                        param_grid=param_test2, scoring='roc_auc', cv=5)
gsearch2.fit(X_train_ppi, y_train_ppi)
print(gsearch2.best_params_)
print(gsearch2.best_score_)
```
